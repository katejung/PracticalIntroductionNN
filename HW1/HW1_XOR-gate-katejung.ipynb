```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Define Hyper-parameters</h2>

```python
gamma = .1 # learning rate
num_iter = 100000
num_hl = 1 # # of hidden layer
num_neuron = 2 # # of neurons in the first (and only) layer
```

<h2>Define your input data and expected labels</h2>

```python
X = np.array([[0,0], [0,1], [1,0],[1,1]], np.float32) # input
Y = np.array([[0],[1],[1],[0]], np.float32) # output
[n,m] = X.shape # n = number of samples, m = number of input features
o = Y.shape[1] # o = number of output features
print(n,m, o)
```

<h2>Define placeholder and Variables</h2>

```python
tf.reset_default_graph()
x = tf.placeholder(tf.float32,shape = [n,m],name = "x") # placeholder for input
y = tf.placeholder(tf.float32,shape = [n,o],name = "y") # placeholder for output
w1 = tf.Variable(tf.random_normal([m,num_neuron]),tf.float32, name = "w1") # variable for the weights of the first layer
w2 = tf.Variable(tf.random_normal([num_neuron,o]),tf.float32, name = "w2") # variable for the weights of the first layer
b1 = tf.Variable(tf.zeros([num_neuron]),tf.float32, name = "b1")
b2 = tf.Variable(tf.zeros([o]),tf.float32, name = "b1")
```

<h2>Define the neural network</h2>

```python
def XOR_gate_1hidden(x):
    x =  tf.cast(x, tf.float32)
    layer1 = tf.sigmoid(tf.matmul(x,w1) + b1)
    y_hat = tf.sigmoid(tf.matmul(layer1, tf.reshape(w2, [num_neuron,o])) + b2)
    return y_hat
```

<h2>Define cost function</h2>

```python
y_hat = XOR_gate_1hidden(x)
cost_func = tf.reduce_mean(-1*(y*tf.log(y_hat)+(1-y)*tf.log(1-y_hat)))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = gamma).minimize(cost_func)
accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.round(y_hat), Y), tf.float32))
```

<h2>Execute training</h2>

```python
#initialize tensorflow variables
init = tf.global_variables_initializer()

#start tensorflow session
with tf.Session() as sess:
    #run initialization
    sess.run(init)
    #training for number of iterations
    for i in range(num_iter):
        #every iteration we run optimization on the training data X and labels Y
        sess.run(optimizer,feed_dict = {x:X,y:Y})
        if i % 10000 == 0:
            print('Epoch ', i)
            print('cost ', sess.run(cost_func, feed_dict={x: X, y: Y}))
            print('accuracy', sess.run(accuracy, feed_dict = {x:X,y:Y}))
    w1 = np.squeeze(sess.run(w1))
    b1 = np.squeeze(sess.run(b1))
    w2 = np.squeeze(sess.run(w2))
    b2 = np.squeeze(sess.run(b2))
```

```python
print(w1, w2, b1, b2)
```

<h2>Plot</h2>

```python
x_min, x_max = np.min(X[:, 0] - 0.2), np.max(X[:, 1]+0.2)
xx, yy= np.meshgrid(np.arange(x_min, x_max, 0.02),np.arange(x_min, x_max, 0.02))
meshInput = np.c_[xx.ravel(), yy.ravel()]
y_hat = tf.Session().run(XOR_gate_1hidden(meshInput))
y_hat = np.round(y_hat)
plt.scatter(meshInput[:,0], meshInput[:,1], c=y_hat, s=10)
plt.scatter(X[:, 0], X[:, 1], c=np.round(np.squeeze(Y)), s=100, cmap='viridis')
print(y_hat)
# plt.xlim([-0.2, 1.2]); 
# plt.ylim([-0.2, 1.25]);
plt.xlabel('x')
plt.ylabel('y')
plt.title('XOR gate using 1 hidden layer')
plt.show()
```
